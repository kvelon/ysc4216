### 1. Generate two classes of data with the same population size (N1=N2=50), sigma = 0.1, mean1 = [2,2], mean2= [3,3]:


```python
def get_data(N1,N2,K,sigma,means) :
    
    # red cluster
    mean = means[0]
    cov = [[sigma[0], 0],[0, sigma[0]]]
    X1 = np.random.multivariate_normal(mean, cov, N1)
    c1 = ['red'] * len(X1)

    # blue cluster
    mean = means[1]
    cov = [[sigma[1], 0],[0, sigma[1]]]
    X2 = np.random.multivariate_normal(mean, cov, N2)
    c2 = ['blue'] * len(X2)
    

    # all points together
    X = np.concatenate((X1,X2))
    color = np.concatenate((c1,c2))

    # T = label 

    T = np.zeros((len(X),1))
    for n in range(0,len(X)) :
        if n >= N1 :
            T[n] = 1

    T = T.astype(int)
    
    return X,T,color


X,T,color = get_data(N1=50,N2=50,K=2,sigma=[0.1,0.1], means=[(2,2),(3,3)])
print(T)
```

### 2. Write code of Gaussian Processes to classify the two clusters.

```python
# kernel function
def kernel(params, xi, xj) :
    k = params[0] * np.exp(-params[1]/2 * np.linalg.norm((xi-xj)**2)) + params[2] + params[3]* np.matmul(xi.T,xj)
    return k 

def kernel_scalar(params,xi,xj) :
    k = params[0] * np.exp(-params[1]/2 * np.linalg.norm((xi-xj)**2)) + params[2] + params[3]* (xi*xj)
    return k 

def sigmoid_func(a) : # a = w.T * x
    M = a.shape[0]
    N = a.shape[1]
    
    res = np.zeros((M, N))
    
    for i in range(M):
        for j in range(N):
            res[i,j] = 1 / (1 + np.exp(-a[i,j]))
    return res

def sig_vec_func(a) :
    return 1 / (1 + np.exp(-a))
```

```python
def get_Cn(params,X,v) :
    
    N = X.shape[0]
    C = np.zeros((N,N))
        
    for i in range(N) :
        # C matrix 
        for j in range(N) :
            if i == j :
                C[i,j] = kernel(params,X[i],X[j]) + v
            else :
                C[i,j] = kernel(params,X[i],X[j])
                
    return C

def get_a_new(params,X,T,v) :
    N = X.shape[0]
    a_old = np.zeros((N,1))
    Cn = get_Cn(params,X,v)
    Cn_inv = np.linalg.inv(Cn)
    
    W = np.zeros((N,N))
    
    for r in range(10) :

        sig_n = sig_vec_func(a_old) # Nx1

        for i in range(N) :
            sig = sig_n[i]
            W[i,i] = sig * (1-sig)
            
        C_W_inv = np.linalg.inv(Cn_inv + W)
        t_s_C_a = T - sig_n - np.matmul(Cn_inv, a_old)

        a_new = a_old + np.matmul(C_W_inv, t_s_C_a)
        a_old = a_new
            
    return Cn, a_new, W
```

```python
def get_testing(params, X, T, v, xstar) :
    C, a_new, W = get_a_new(params,X,T,v)
    
    N = X.shape[0]
    
    W_inv = np.linalg.inv(W)
    C_inv = np.linalg.inv(C)
    
    # kstar
    xstar = np.reshape(xstar,(2,1))
    kstar = np.reshape([kernel(params,X[i],xstar) for i in range(N)], (N,1)) # Nx1    
    
    # cstar
    cstar = kernel(params, xstar, xstar) + v # 1x1

    # mu, sigma star
    mustar = np.matmul(kstar.T, np.matmul(C_inv,a_new))
    sstar = cstar - np.matmul(kstar.T, np.matmul(np.linalg.inv(W_inv + C), kstar))
    
    # prob with sigmoid
    frac = mustar / (1 + np.pi/8 * sstar**2) ** (1/2)
    prob = sig_vec_func(frac)
    
    return prob 
```

### 6. Discuss why the new parameters generate a less precise probability map.


Old parameters: (1.0, 4.0, 1.0, 2.0)


New parameters: (1.0, 0.25, 0, 0)


The old parameters are all non-zero values, whereas the new parameters are 0 for the 3rd and 4th parameters. This means that the old kernel generally takes into account more information about the x inputs. 

Specifically, the kernel equation is: $\theta_0 exp(-\frac{\theta_1}{2} ||x_i-x_j||^2) + \theta_2 + \theta_3 x_i^Tx_j$. This means that for the new parameters, the kernel equation actually simplifies to $\theta_0 exp(-\frac{\theta_1}{2} ||x_i-x_j||^2)$. Thus, information about the x input embedded into $x_i^Tx_j$ is not taken into account when calculating $c_*$, which results in a less precise probability map.

### 10. Change the means and clusters back to #1 above, and add another cluster belonging to C2, with mean=[1,1] and sigma=0.1, like: 


```python
def get_data_out(N1,N2,N3,K,sigma,means) :
    
    # red cluster
    mean = means[0]
    cov = [[sigma, 0],[0, sigma]]
    X1 = np.random.multivariate_normal(mean, cov, N1)
    c1 = ['red'] * len(X1)

    # blue cluster
    mean = means[1]
    cov = [[sigma, 0],[0, sigma]]
    X2 = np.random.multivariate_normal(mean, cov, N2)
    c2 = ['blue'] * len(X2)
    
    # blue outliers 
    mean = means[2]
    cov = [[sigma, 0],[0, sigma]]
    X3 = np.random.multivariate_normal(mean, cov, N2)
    c3 = ['blue'] * len(X3)

    # all points together
    X = np.concatenate((X1,X2,X3))
    color = np.concatenate((c1,c2,c3))

    # T = label 
    T = np.zeros([len(X),1])
    for n in range(0,len(X)) :
        if n >= N1 and n < len(X1) + len(X2) + len(X3):
            T[n] = 1

    T = T.astype(int)
    
    
    return X,T,color


X3,T3,color3 = get_data_out(N1=50,N2=50,N3=50,K=2,sigma=0.1,means=[(2,2),(3,3),(1,1)])


plt.scatter(X3[:,0], X3[:,1], marker = 'o', c = color3, edgecolor = 'black')

plt.show()
```
